## 3° Pré-Processamento dos Dados

```
# Eliminar a coluna de identificação, pois ela não é variável explicativa, é somente a identificação do cliente7
dataset = dataset.drop(columns='id_cliente')
```

```
# Separando as Variáveis de Entrada e Saída do Modelo
X = dataset.iloc[:,:-1].values  # Variáveis independentes
y = dataset.iloc[:,-1].values   # Target/labe/class/vairável dependente
```

```
# Fazendo o Print das variáveis X e Y
print("\nVariáveis de Entrada da avaliação:\n\n", X[:3])
print("\nRisco - Variável a ser Predita:\n\n", y[:3])
```

```
# Vamos verificar o "X"
X
```

```
# Vamos verificar o "Y"
y
```

## Separando os dados entre Treino e Teste

```
# Divide o dataset em Train and Test 
# Vamos considerar " 20% " dos dados para teste e " 80% " dos dados para treino
# Vamos usar essa função "TRAIN_TES_SPLIT"
X_train, X_test, y_train, y_test = train_test_split(X ,y, test_size=0.2, random_state = 7) 
```

```
# Vamos fazer o print
print("Dados de Treino :\n----------------\n")
print("X = \n", X_train[:2])
print("y = \n", y_train[:2])

print("\n\nDados de Teste :\n----------------\n")
print("X = \n",X_test[:2])
print("y = \n", y_test[:2])
```

```
# Vamos ver o tamanho da nossa amostra que foi feita essa divisão. 
print(f"Shape dos dados de treino é {X_train.shape}")
print(f"Shape dos dados de teste é {X_test.shape}")
```

## Padronizando as Features - As variáveis explicativas (*variáveis de entrada*)

```
# Redimensionando os dados - PADRONIZAÇÂO - Feature scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()

X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test) 

print("\n--------------------------------\nDataset depois da padronização:\n--------------------------------\n", )

print("\nX_train :\n", X_train[:5])
print("-------------------------")
print("\nX_test :\n", X_test[:5])
```

## 4° Construindo a Máquina Preditiva

## 1° Versão da Máquina Preditiva: Criando um **baseline**

- Aqui nós vamos criar a primeira versão da Máquina Preditiva. 

- Depois, vamos buscar formas de melhorar a assertividade do modelo.

### ***Utilizaremos o método KNN***

*Eese métodos é considerado como o **algoritmo do vizinho mais próximo** e requer que:*

- Os dados de treinamento devem estar **normalizados** (*ou padronizados*)

- Deve-se definir uma **métrica** para calcular a distância (*no nosso caso será a distância euclidiana*) 

- Deve-se definir o **valor** de **K** (*ou seja, o número de vizinhos mais próximos que serão considerados pelo algoritmo*).


## *A distância euclidiana é calcula pela formula abaixo*


image.png

```
# KNN ->> é conhecido como o algoritimo do vizinho mais proximo.
# KNN ->> Cálcula a distancia entre os vizinhos mais proximos referente ao registro que desejamos descobrir.
# A distancia é calculada usando os conceitos basico da mtemática (Como a distancia entre 2 pontos)
# Nesse metodo também definimos a quantidade de "K->> K representa os vizinhos mais proximos"(k=3-> 3 vizinhos mais proximos)

# Importando a Biblioteca KNN
from sklearn.neighbors import KNeighborsClassifier
Maquina_Preditiva = KNeighborsClassifier(n_neighbors=30) 
# Escolhemos "k =30", lembrando que você pode escolher os valores [1, 10, 30, 50]
```

```
# Treinando a Máquina preditiva com os dados de treino
# A funçao ".fit" faz o treinamento da nossa máquina e depois passamos pra ela os dados de treinos.
Maquina_Preditiva = Maquina_Preditiva.fit(X_train,y_train)
```

```
# Fazendo novas Previsões com os dados de teste
y_pred = Maquina_Preditiva.predict(X_test)


print("\n---------------------------\n")
print("Valores Reais com dados de teste :\n",y_test)

print("\n---------------------------\n")
print("Valores Preditos com a Máquina Preditiva com dados de Teste :\n",y_pred)
```

```
# Função que mostra os hyperparâmetros utilizados na construção da Máquina Preditiva 
# Essa função "get_params"->> ela nos mostra quais foram os hyperparâmetros utilizados na construção da nossa
# máquina preditiva
Maquina_Preditiva.get_params
```

## 5° Avaliando a Máquina Preditiva

```
# Calculando a Acurácia
# A Acurácia é cálculada pela razão entre o "Total de acertos pelo Total Geral"
from sklearn import metrics
print("Acurácia da Máquina = ", metrics.accuracy_score(y_test, y_pred)*100)
```

```
# Comparando Real e Predito com dados de Teste
# Criando o outro DataFrame com o valor Real e Previsto afim verificarmos o comportamento da nossa máquina
print("\nReal vs Predito \n------------------------------\n")
error_df = pd.DataFrame({"Real" : y_test,
                         "Predito" : y_pred})

error_df.head(5)
```

```
# Usando a confusion matrix  e o classification report para avaliar a acurácia e demais indicadores
from sklearn.metrics import confusion_matrix, classification_report
from sklearn import metrics

print("Acurácia da Máquina = ", metrics.accuracy_score(y_test, y_pred)*100)
cm = confusion_matrix(y_test,y_pred)
cr = classification_report(y_test, y_pred)
print(cm)
print(cr)

# A "confusion_metrix"->> nos mostra alguns parêmetros, como a acurácia entre outros.
# Na matriz (que é a confusion_metrix) temos que a diagonal principal não nula 
# Essa diagonal é chamada de diagonal do bem -> representa todos os acertos que o nosso modelo 
# E os demais representam os erros de modelos 
```

# 2° Versão da Máquina Preditiva: **Máquina Definitiva**

### *Vamos utilisar o Support Vector Machines* - **SVM**

- O VSM - Faz a separação usando o vetor de suporte afim de cálcular a melhor ou maior distância para obter a melhor divisão entre os pontos



image.png

### ***Etapas para o método SVM***

- Definir o Kernel (*o Kernel é uma formula matemática de fazer essas divisão*).

- Entendimento do Hiperplano

- Hiperparâmetros (começar a profundar)

- Técnicas de Balanceamento de Classe

- Novas Métricas de Avaliação dos Algoritmos AUC


image.png

```
# Importando a Biblioteca do SVM
# Vale lembrar que dentro do "SVM"temos dois tipos de algoritmos que são:
# ** SVC->> Support Vetor Classifiers ->> Usado para prever uma classe
# ** SVR ->> Support Vetor Regression ->> Regressão é quando a gente quer prever um valor
# No nosso caso vamos usar o SVC, pois queremos prever uma classe.

from sklearn.svm import SVC
classifier = SVC(kernel='linear',             # ['linear', 'rbf']
                 gamma= 1e-5  ,               # [1e-1, 1e-2, 1e-3, 1e-5]
                 C=10,                        # [10.0, 100.0, 1000.0, 10000.0]
                 random_state=7)

# O "gama" e "C" são parâmetros de regularização que visão melhorar a acurácia
```

```
# Treinando a Máquina com o SVC
classifier.fit(X_train,y_train)
```

```
# Fazendo novas previsões com os dados de teste
y_pred = classifier.predict(X_test)

print("\n---------------------------\n")
print("Predicted Values for Test Set :\n",y_pred)
print("\n---------------------------\n")
print("Actual Values for Test Set :\n",y_test)
```

```
# Imprimindo os hyperparâmetros utilizados na criação da Maquina Preditiva Definitiva
classifier
```

```
# Calculando a Acurácia da Máquina Preditiva
from sklearn import metrics
print("Prediction Accuracy = ", metrics.accuracy_score(y_test, y_pred))

# Comparando valores Reais com Preditos pela Máquina
print("\nActual vs Predicted  \n------------------------------\n")
error_df = pd.DataFrame({"Actual" : y_test,
                         "Predicted" : y_pred})

error_df.head(5)
```

```
# Usando a confusion matrix e o classification report para avaliar a Máquina
from sklearn.metrics import confusion_matrix, classification_report

cm = confusion_matrix(y_test,y_pred)
cr = classification_report(y_test, y_pred)
```

```
# Imprimindo a Confusion Matrix e o Classification Report
print(cm)
print(cr)
```
